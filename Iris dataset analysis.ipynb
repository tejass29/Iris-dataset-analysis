```python
#importing the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes = True)
%matplotlib inline
```

```python
#imorting Iris dataset
data = pd.read_csv("D:\Projects !\Iris.csv")
data.head()
```

```python
#knowing the dataset 
print(data.shape)
```

```python
print(data.info())
```

```python
print(data.groupby("Species").size())
```

```python
#visualising relations between variables based on species
plt.figure(figsize = (16,9))
plt.title("Comparision between sepal width and sepal length")
sns.scatterplot(data["SepalLengthCm"], data["SepalWidthCm"], hue = data["Species"], s = 80)
```

```python
#visualising length and width of petal and sepal through histogram
data.hist(edgecolor = "white", linewidth = 1.2)
```

```python
#visualising petal length for each species
sns.violinplot(data=data, x = 'Species', y = 'PetalLengthCm')
```

```python
#visualising all columns relationship using pairplots.
sns.pairplot(data, hue = "Species", height = 3)
```
